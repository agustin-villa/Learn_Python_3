```python
def contains(big_string, little_string):
  if little_string in big_string:
    return True
  else:
    return False

# Write a function that makes a list of char in common, list should be unique char
def common_letters(string_one, string_two):
  common = []
  for char in string_one:
    if char in string_two and char not in common:
      common.append(char)
  
  return common
```

```python
# Create function username_generator
def username_generator(first_name, last_name):
  if len(first_name) < 3 or len(last_name) < 4:
    return first_name+last_name
  else:
    return first_name[:3] + last_name[:4]

# Create function password_generator that shifts the user_name to the left one to create the password 
def password_generator(user_name):
  password = ''
  for character in range(len(user_name)):
    password+=(user_name[character-1])
  return password
```
